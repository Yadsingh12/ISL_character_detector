```python
import os
import shutil
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img, array_to_img
```

```python
# Path to the dataset
base_path = '/content/drive/My Drive/Colab Notebooks/ISLtoText/ISL_Dataset2'
```

```python
def count_images_in_folders(path):
    counts = {}
    for class_name in os.listdir(path):
        class_path = os.path.join(path, class_name)
        if os.path.isdir(class_path):
            counts[class_name] = len(os.listdir(class_path))
    return counts
```

```python
# Count the images
image_counts = count_images_in_folders(base_path)

# Print the counts
for class_name, count in image_counts.items():
    print(f'Class {class_name}: {count} images')
```

```python
def rename_images_in_folders(path):
    for class_name in os.listdir(path):
        class_path = os.path.join(path, class_name)
        if os.path.isdir(class_path):
            for i, filename in enumerate(os.listdir(class_path)):
                file_extension = os.path.splitext(filename)[1]
                new_name = f'{class_name}_{i+1}{file_extension}'
                old_file_path = os.path.join(class_path, filename)
                new_file_path = os.path.join(class_path, new_name)
                os.rename(old_file_path, new_file_path)
```

```python

# Rename the images
rename_images_in_folders(base_path)

# Verify the renaming
for class_name in os.listdir(base_path):
    class_path = os.path.join(base_path, class_name)
    if os.path.isdir(class_path):
        print(f'Images in class {class_name}:')
        for filename in os.listdir(class_path):
            print(filename)
        print('\n')
```

```python
augmented_data_path = '/content/drive/My Drive/Colab Notebooks/ISLtoText/ISL_Augmented_Dataset2'
```

```python

target_count = 800  # Target number of images per class after augmentation

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

def augment_images(class_name):
    class_path = os.path.join(base_path, class_name)
    augmented_class_path = os.path.join(augmented_data_path, class_name)
    os.makedirs(augmented_class_path, exist_ok=True)

    images = os.listdir(class_path)
    current_count = len(images)

    # Copy original images to the augmented dataset directory
    for image_name in images:
        src = os.path.join(class_path, image_name)
        dst = os.path.join(augmented_class_path, image_name)
        shutil.copy(src, dst)

    # If not enough images, augment them
    while current_count < target_count:
        for i, image_name in enumerate(images):
            img_path = os.path.join(class_path, image_name)
            img = load_img(img_path)
            x = img_to_array(img)
            x = x.reshape((1,) + x.shape)

            # Create a generator to augment the image
            generator = datagen.flow(x, batch_size=1)

            # Generate a single augmented image
            batch = next(generator)
            augmented_img_name = f"{class_name}_{current_count + 1}.jpeg"
            augmented_img_path = os.path.join(augmented_class_path, augmented_img_name)
            img_to_save = array_to_img(batch[0])
            img_to_save.save(augmented_img_path)

            current_count += 1
            if current_count >= target_count:
                print(f"{class_name} completed")
                break

# Augment images for each class
for class_name in os.listdir(base_path):
    if os.path.isdir(os.path.join(base_path, class_name)):
        augment_images(class_name)

```

```python
# Count the images
image_counts = count_images_in_folders(augmented_data_path)

# Print the counts
for class_name, count in image_counts.items():
    print(f'Class {class_name}: {count} images')
```
